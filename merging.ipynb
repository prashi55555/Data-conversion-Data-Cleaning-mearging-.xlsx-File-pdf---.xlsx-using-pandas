```python
import pandas as pd 
file2 = "F:/8th sem/Cleaning of Data & Merging into single excel/Data Analyst Quick Assignment.xlsx"
```

```python
import pandas as pd
from openpyxl import load_workbook
fileSummary = "F:/8th sem/Cleaning of Data & Merging into single excel/cleaned payout & summary files/summary.xlsx files/summary.xlsx"
df = pd.read_excel(fileSummary, sheet_name=0)

df1 = pd.read_excel(file2 ,sheet_name ='Summary')
df2 = pd.concat([df,df1], ignore_index = True)
with pd.ExcelWriter(file2, engine="openpyxl", mode='a', if_sheet_exists='overlay') as writer:
    df2.to_excel(writer, sheet_name='Summary', index=False)
```
