```python
import pdfplumber
import pandas as pd

pf1 = "F:/8th sem/Cleaning of Data & Merging into single excel/Commission Invoices/taco_Tax_Invoice_600278_09042025_250409FS29010197.pdf"
ef1 = "commission_600278.xlsx"

with pd.ExcelWriter(ef1, engine='openpyxl') as writer:
    sheet_number = 1
    with pdfplumber.open(pf1) as pdf:
        for page in pdf.pages:
            tables = page.extract_tables()
            for table in tables:
                df = pd.DataFrame(table)
                sheet_name = f"Sheet{sheet_number}"
                df.to_excel(writer, index=False, header=False, sheet_name=sheet_name)
                sheet_number += 1
```
